```python
import pandas as pd
import numpy as np

df = pd.read_csv("NVH_Data_Synthetic")
display(df)
```

```python
# Identifying all object-type columns
print("\nColumns with 'object' dtype (possible mixed types):")
print(df.select_dtypes(include='object').columns)

#U nique entries in suspect numeric columns
print("\nUnique 'Speed_kmph' entries:")
print(df['Speed_kmph'].unique())
print("\nUnique 'Noise_dB' entries:")
print(df['Noise_dB'].unique())
print("\nUnique 'Vibration_magnitude' entries:")
print(df['Vibration_magnitude'].unique())
print("\nUnique 'Road_Surface' entries:")
print(df['Road_Surface'].unique())
```

```python
#Removing Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Replace's text-based errors in numeric columns with NaN
df['Speed_kmph'] = pd.to_numeric(df['Speed_kmph'], errors='coerce')
df['Noise_dB'] = pd.to_numeric(df['Noise_dB'], errors='coerce')
df['Vibration_magnitude'] = pd.to_numeric(df['Vibration_magnitude'], errors='coerce')
df['Customer_Feedback_Score'] = pd.to_numeric(df['Customer_Feedback_Score'], errors='coerce')
```

```python
#Filtering Out Unrealistic or Out-of-Range Values
df = df[(df['Speed_kmph'] >= 0) & (df['Speed_kmph'] <= 140)]
df = df[(df['Noise_dB'] >= 40) & (df['Noise_dB'] <= 120)]
df = df[(df['Vibration_magnitude'] >= 0) & (df['Vibration_magnitude'] <= 10)]
df = df[df['Customer_Feedback_Score'].between(1, 5)]
```

```python
# Fixing Road_Surface inconsistencies
df['Road_Surface'] = df['Road_Surface'].str.strip().str.title().replace({
    'Ashpalt': 'Asphalt', 'AsPhalt': 'Asphalt',
    'Gravle': 'Gravel', 'Gravel ': 'Gravel',
    'Wet Asphalt ': 'Wet Asphalt'
})

# Fixing Component_Source variations
df['Component_Source'] = df['Component_Source'].str.strip().str.title().replace({
    'Engine ': 'Engine', 'Engin': 'Engine', 'Enginee': 'Engine',
    'Tyres': 'Tires', 'Tire': 'Tires'
})

# Fixing Vehicle_Model
df['Vehicle_Model'] = df['Vehicle_Model'].str.strip().replace({
    'Modle A': 'Model A', 'Model A ': 'Model A', 'Model_B': 'Model B',
    'model A': 'Model A', 'Mod. B': 'Model B'
})
```

```python

# Fill"s missing numerics with column median
df['Noise_dB'].fillna(df['Noise_dB'].median(), inplace=True)
df['Speed_kmph'].fillna(df['Speed_kmph'].median(), inplace=True)
df['Vibration_magnitude'].fillna(df['Vibration_magnitude'].median(), inplace=True)
df['Customer_Feedback_Score'].fillna(df['Customer_Feedback_Score'].mode()[0], inplace=True)
df['Operating_Temperature'].fillna(df['Operating_Temperature'].median(), inplace=True)

# Replacing missing categorical text fields with placeholders
df['Road_Surface'].fillna("Unknown", inplace=True)
df['Component_Source'].fillna("Unknown", inplace=True)
df['Measurement_Point'].fillna("Unknown", inplace=True)

# Replacing missing feedback comments
df['Feedback_Comment'].fillna("No Feedback Provided", inplace=True)

# Drop rows with critical missing identifiers
df.dropna(subset=['Timestamp', 'Vehicle_Model', 'Manufacture_Date'], inplace=True)
```

```python
#Checking for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Noise Category
df['Noise_Category'] = df['Noise_dB'].apply(
    lambda x: 'Low Noise' if x < 60 else ('Moderate Noise' if x < 80 else 'High Noise')
)

# Speed Range
df['Speed_Range'] = df['Speed_kmph'].apply(
    lambda x: 'Low Speed' if x < 40 else ('Medium Speed' if x < 80 else 'High Speed')
)

# Harshness Rating (basic sentiment logic)
def rate_ride(score, comment):
    comment = str(comment).lower()
    if score >= 4 or 'smooth' in comment or 'excellent' in comment or 'quiet' in comment:
        return 'Good Ride'
    elif score <= 2 or 'uncomfortable' in comment or 'vibrations' in comment or 'rattling' in comment:
        return 'Poor Ride'
    else:
        return 'Average Ride'

df['Harshness_Rating'] = df.apply(lambda row: rate_ride(row['Customer_Feedback_Score'], row['Feedback_Comment']), axis=1)
```

```python
#Creating Dimension Tables for Star Schema
# Dim_Vehicle
dim_vehicle = df[['Vehicle_Model', 'Manufacture_Date']].drop_duplicates().reset_index(drop=True)
dim_vehicle['Vehicle_SK'] = dim_vehicle.index + 1

# Dim_Location
dim_location = df[['Road_Surface']].drop_duplicates().reset_index(drop=True)
dim_location['Location_SK'] = dim_location.index + 1

# Dim_Component
dim_component = df[['Component_Source', 'Measurement_Point']].drop_duplicates().reset_index(drop=True)
dim_component['Component_SK'] = dim_component.index + 1
```

```python
#Merging Surrogate keys to fact table
df = df.merge(dim_vehicle, on=['Vehicle_Model', 'Manufacture_Date'], how='left')
df = df.merge(dim_location, on='Road_Surface', how='left')
df = df.merge(dim_component, on=['Component_Source', 'Measurement_Point'], how='left')
```

```python
df_fact = df.drop(columns=['Vehicle_Model', 'Manufacture_Date', 'Road_Surface', 'Component_Source', 'Measurement_Point'])
```
